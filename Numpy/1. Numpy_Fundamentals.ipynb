## Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)

print(a1)
print()
print(a2)
```

```python
# scalar operations

# arithmetic
a1 ** 2
```

```python
# relational
a2 == 15
```

```python
# vector operations
# arithmetic
a1 ** a2
```

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row
np.max(a1,axis=0)
```

```python
# mean/median/std/var
np.var(a1,axis=1)
```

```python
# trigonomoetric functions
np.sin(a1)
```

```python
# dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponents
np.exp(a1)
```

```python
# round/floor/ceil

print(np.round(np.random.random((2,3))*100))
print()
print(np.floor(np.random.random((2,3))*100))
print()
print(np.ceil(np.random.random((2,3))*100))
```
